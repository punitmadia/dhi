```python
t = np.arange(0,10,0.5) # Evenly sampled time at 0.5 second intervals

plt.plot(t,t,'b--') # y = x: dashed blue
plt.plot(t,t**2,'rs') # y = x^2: red squares
plt.plot(t,t**3,'g-')
```
